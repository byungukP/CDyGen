```python
import numpy as np

# RMSD xvg file loading

rmsd_xvg_dir = "/mnt/e/masif_docker/pmp/htmd/output_files/traj_fit_test/2ADZ_A/prod/rmsd_check"
xvg_file = "rmsd.xvg"

def read_xvg(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    data = []
    for line in lines:
        if not line.startswith(('#', '@')):
            data.append([float(x) for x in line.split()])
    
    return np.array(data)

rmsd_data = read_xvg(f"{rmsd_xvg_dir}/{xvg_file}")
```

```python
# plotting
plt.plot(rmsd_data[:, 0], rmsd_data[:, 1])
plt.xlim(0, 11000)
plt.xlabel('Time [ps]')
plt.ylabel('RMSD [nm]')
plt.title('RMSD over MD trajectory')
# plt.grid(True)
plt.show()
```
